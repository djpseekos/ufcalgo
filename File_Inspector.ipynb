```python

import pandas as pd
from pathlib import Path
```

```python
DATA_DIR = Path("data/curated")

files = [
    "events.csv",
    "fights.csv",
    "fighters.csv",
    "stats_round.csv",
    "odds_snapshot.csv",
    "fighter_event.csv"
]
```

```python
# koitaei kathe stylh gia to ti einai
for f in files:
    path = DATA_DIR / f #koitaei ola sto data curated
    if path.exists():
        print(f"\n--- {f} ---")
        df = pd.read_csv(path)
        print(df.dtypes) 
        print(df.head(3))      
        print(f"Shape: {df.shape}")
    else:
        print(f"\n!!! {f} not found")
```
